#### Distancia Haversine

```python
def haversine_distance(lon1, lat1, lon2, lat2):
    R = 6371  # Radio de la Tierra en kilómetros
    lat1_rad = radians(lat1)
    lon1_rad = radians(lon1)
    lat2_rad = radians(lat2)
    lon2_rad = radians(lon2)
    dlon = lon2_rad - lon1_rad
    dlat = lat2_rad - lat1_rad
    # Fórmula Haversine
    a = sin(dlat / 2)**2 + cos(lat1_rad) * cos(lat2_rad) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance # en km
```

```python
def inferir_descensos_por_id(smart_instancia, stops_instancia, resultado):
    resultados = []
    no_procesados = []
    i = 0
    n = len(smart_instancia)
    while i < n:
        # Último registro (no tiene pareja)
        if i == n - 1:
            no_procesados.append({
                'indice': i,
                'UsuarioID': smart_instancia.iloc[i]['CodigoTarjeta'],
                'Razon': 'Registro único'
            })
            break
            
        usuario_actual = smart_instancia.iloc[i]['CodigoTarjeta']
        usuario_siguiente = smart_instancia.iloc[i+1]['CodigoTarjeta']
        
        # Caso par ida-vuelta
        if usuario_actual == usuario_siguiente:
            try:
                # Datos IDA
                id_linea_ida = smart_instancia.iloc[i]['CodigoLinea']
                codigo_parada_ida = smart_instancia.iloc[i]['CodigoParada']
                
                # Datos VUELTA (posible descenso IDA)
                codigo_parada_vuelta = smart_instancia.iloc[i+1]['CodigoParada']
                
                # Verificar si la parada de vuelta existe en stops_instancia
                mask = (stops_instancia['ID-PARADA'] == codigo_parada_vuelta)
                if not mask.any():
                    raise ValueError(f"Parada {codigo_parada_vuelta} no encontrada")
                
                # Obtener paradas cercanas desde resultado
                paradas_cercanas = resultado.loc[
                    resultado['ID-PARADA'] == codigo_parada_vuelta, 'ID-PARADA_2'
                ].values[0]
                
                # Filtrar paradas que pertenezcan a la misma línea
                paradas_validas = []
                for parada_id in paradas_cercanas:
                    # Verificar si la parada existe y tiene la misma línea
                    mask_parada = (
                        (stops_instancia['ID-PARADA'] == parada_id) & 
                        (stops_instancia['ID-LINEA'] == id_linea_ida)
                    )
                    if mask_parada.any():
                        paradas_validas.append(parada_id)
                
                if paradas_validas:
                    resultados.append({
                        'UsuarioID': usuario_actual,
                        'Tipo': 'IDA',
                        'ID_Linea': id_linea_ida,
                        'ID_Parada_Ascenso': codigo_parada_ida,
                        'ID_Parada_Descenso_Estimado': paradas_validas[0],
                        'ID_Parada_Descenso_Real': codigo_parada_vuelta,
                        'Coincidencia_Linea': True,
                        'Estatus': 'EXITO'
                    })
                

                
                i += 2
                
            except Exception as e:
                resultados.append({
                    'UsuarioID': usuario_actual,
                    'Estatus': 'FALLIDO',
                    'Error': str(e)
                })
                i += 1
        else:
            no_procesados.append({
                'indice': i,
                'UsuarioID': usuario_actual,
                'Razon': 'Usuario diferente'
            })
            i += 1
    
    return {
        'descensos_inferidos': pd.DataFrame(resultados),
        'no_procesados': pd.DataFrame(no_procesados)
    }
```

```python
resultados_inferencia = []
# Verificamos que el DataFrame no esté vacío
if len(smart_instancia) == 0:
    raise ValueError("El DataFrame smart_instancia está vacío")

# Verificamos que haya un número par de registros
if len(smart_instancia) % 2 != 0:
    print("Advertencia: Número impar de registros. El último registro no se procesará")

num_usuarios = len(smart_instancia) // 2

for i in range(num_usuarios):
    try:
        idx_ida = i * 2
        idx_vuelta = i * 2 + 1
        
        # Procesamiento IDA --------------------------------------------
        linea_ida = smart_instancia["NombreLinea"].iloc[idx_ida]
        parada_ida = smart_instancia["NombreParada"].iloc[idx_ida]
        nuevo_ascenso_ida = smart_instancia["NombreParada"].iloc[idx_vuelta]
        usuario = smart_instancia['CodigoTarjeta'].iloc[idx_ida]
        fecha_hora_ida = smart_instancia["FechaHoraValidacion"].iloc[idx_ida]
        fecha_hora_regreso = smart_instancia["FechaHoraValidacion"].iloc[idx_vuelta]
        # Buscar id_ascenso con manejo de error
        try:
            id_ascenso_ida = stops_instancia.loc[
                stops_instancia['nombres'] == nuevo_ascenso_ida, 'id2'
            ].iloc[0]
        except IndexError:
            #print(f"No se encontró la parada {nuevo_ascenso_ida} en stops_instancia")
            continue
            
        # Resto del procesamiento IDA
        paradas_cercanas_ida = resultado.loc[resultado['id2'] == id_ascenso_ida, 'id2_2'].values
        if len(paradas_cercanas_ida) == 0:
            continue
        paradas_cercanas_ida = paradas_cercanas_ida[0]
        
        lineas_parada_ida = stops_instancia.loc[
            stops_instancia['id2'] == id_ascenso_ida, 'bus_lines'
        ].values
        if len(lineas_parada_ida) == 0:
            continue
        lineas_parada_ida = lineas_parada_ida[0]
        
        paradas_comunes_ida = []
        for parada_id in paradas_cercanas_ida:
            fila = stops_instancia.loc[stops_instancia['id2'] == parada_id, 'bus_lines']
            if not fila.empty:
                lineas_paradas = fila.iloc[0]
                if any(linea in lineas_paradas for linea in lineas_parada_ida):
                    paradas_comunes_ida.append(parada_id)
        
        if paradas_comunes_ida:
            parada_descenso_ida = paradas_comunes_ida[0]
            #print(f"Usuario {usuario}: Descenso IDA inferido en {parada_descenso_ida}")
            resultados_inferencia.append({
            'CodigoTarjeta': usuario,
            'FechaHoraAscenso': fecha_hora_ida,
            'ParadaAscenso_Nombre': parada_ida,
            'LineaViaje': linea_ida,
            'ParadaRegreso': nuevo_ascenso_ida,
            'FechaHoraRegreso': fecha_hora_regreso,
            'ParadaDescenso_Inferida': parada_descenso_ida # El resultado de la inferencia (o None)
            })
        
    except Exception as e:
        print(f"Error procesando usuario {i}: {str(e)}")
        continue
```
